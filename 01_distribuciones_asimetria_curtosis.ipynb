```python
from utils_distribuciones import generar_distribucion
import numpy as np
```

```python
mu = 0
sigma = 1
dist_normal = np.random.normal(mu,sigma,1000)
```

```python
# Uso de nuestra funcion generica
generar_distribucion(dist_normal, 'Distribuci√≥n normal')
```

# Distribuci√≥n sesgada a la derecha

```python
dist_sesgada = np.random.exponential(1,1000)
generar_distribucion(dist_sesgada, 'Distribuci√≥n sesgada a la derecha')
```

# Distribuci√≥in sesgada a la izquierda

```python
dist_sesgada_izquierda = -np.random.exponential(1,1000)
generar_distribucion(dist_sesgada_izquierda, 'Distribuci√≥n sesgada a la izquierda')
```

# Distribuci√≥n Bimodal

```python
dist_bimodal = np.concatenate(
    [np.random.normal(mu,sigma,1000),
    np.random.normal(mu,sigma,1000)]
)
generar_distribucion(dist_bimodal, 'Distribuci√≥n bimodal')
```

# Distribuci√≥n uniforme

```python
dist_uniforme = np.random.uniform(-1, 1, 1000)

# Uso de nuestra funcion generica
generar_distribucion(dist_uniforme, 'Distribucion Uniforme')
```
