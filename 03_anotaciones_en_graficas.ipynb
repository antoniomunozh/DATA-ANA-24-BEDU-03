```python
import pandas as pd
import matplotlib.pyplot as plt
# Crear los datos
data = {
    'Mes': ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre'],
    'Ventas': [150, 200, 250, 300, 280, 320, 310, 330, 340, 360, 380, 400]
}

# Crear un DataFrame
datos = pd.DataFrame(data)
datos.head()
```

```python
# Grafica de barras
plt.figure(figsize=(10,6))
plt.bar(datos['Mes'], datos['Ventas'], color='skyblue')
```

```python
# Grafica de barras con titulo
plt.figure(figsize=(10,6))
plt.bar(datos['Mes'], datos['Ventas'], color='skyblue')
plt.title('Ventas mensuales')
plt.xlabel('Mes de la venta')
plt.ylabel('Ventas')
plt.xticks(rotation=20)
```

```python
# Grafica de barras con titulo
plt.figure(figsize=(10,6))
plt.bar(datos['Mes'], datos['Ventas'], color='skyblue')
plt.title('Ventas mensuales')
plt.xlabel('Mes de la venta')
plt.ylabel('Ventas')
plt.xticks(rotation=20)
plt.grid(True, which='both', linestyle='--', linewidth=0.4)
```

```python
# Configurar grafica con titulo
plt.figure(figsize=(10, 6))
plt.bar(datos['Mes'], datos['Ventas'], color='skyblue')

# titulo y etiquetas
plt.title('Ventas mensuales')
plt.xlabel('Mes de la venta')
plt.ylabel('Ventas')

# Rotar etiquetas
plt.xticks(rotation=45)
plt.yticks(rotation=45)
# Mostrar grid
plt.grid(True, which='both', linestyle='--', linewidth=0.4)

# Resaltar mes con mayor venta
max_ventas = datos['Ventas'].max()
mes_max_ventas = datos['Mes'][datos['Ventas'].idxmax()]

# Agregar anotacion
plt.annotate(
    'Mayor venta',
    xy=(mes_max_ventas, max_ventas),
    xytext=(mes_max_ventas, max_ventas + 40),
    arrowprops=dict(facecolor='red', shrink=0.05))
```
